Decision Tree

Survival rate of titanic passengers based on class, sex, age and ticket fare

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv('titanic.csv')
print(len(df))
df.head()
```

```python
data = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
data.head()
```

```python
input = data.drop('Survived', axis='columns')
target = data[['Survived']]
input.head(10)
```

```python
Sex_e = LabelEncoder()
input['Sex'] = input.Sex.map({'male':1, 'female':2})
input.head(10)
```

```python
input.Age = input.Age.fillna(math.floor(input.Age.mean()))
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2)
len(X_train)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.predict(X_test.head())
```
